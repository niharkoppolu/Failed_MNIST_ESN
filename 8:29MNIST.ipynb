```python
#This is the Current 8/29 Reservoir Computer
import numpy
import scipy.special #used for sigmoid function
import matplotlib.pyplot
%matplotlib inline
```

```python
#1. Create function that processes MNIST data into training and testing data sets

def MNIST_process_data():
    #A. Read in training data, create x_train and y_train
    #_________________________________________________________________________________________________________________________
    tr_data_file = open(r"/Users/niharkoppolu/Desktop/Number_Detection_Project/mnist_data_set/mnist_train.csv")
    tr_data_list = tr_data_file.readlines()
    tr_data_file.close()

    index = 0
    x_train = []
    y_train = []
    for record in tr_data_list:

        #convert test data from list off comma sep. vals to array of floats
        all_values = record.split(',')

        #rescale data from 0 - 255 to range btw 0 - 0.99
        x_train.append((numpy.asfarray(all_values[1:]) / 255.0 * 0.99) + 0.01)

        #get y_train
        #y_train.append(numpy.asfarray(all_values[0]))
        y_train.append(int(all_values[0]))

        index += 1
        pass
    #__________________________________________________________________________________________________________________________
    


    #__________________________________________________________________________________________________________________________
    #B. Reading in TESTING DATA: x_test, y_test 
    test_data_file = open(r"/Users/niharkoppolu/Desktop/Number_Detection_Project/mnist_data_set/mnist_test.csv")
    test_data_list = test_data_file.readlines()
    test_data_file.close()

    index = 0
    x_test = []
    y_test = []
    for record in test_data_list:

        #convert test data from list off comma sep. vals to array of floats
        all_values = record.split(',')

        #rescale data from 0 - 255 to range btw 0 - 0.99
        x_test.append((numpy.asfarray(all_values[1:]) / 255.0 * 0.99) + 0.01)

        #get y_test
        #y_test.append(numpy.asfarray(all_values[0]))
        y_test.append(int(all_values[0]))

        index += 1
        pass
    
    #Test x_train and y_train: (WORKS PERFECTLY)
    image_array = ((x_test[5] - 0.01) / 0.99 * 255.0).reshape((28,28))
    matplotlib.pyplot.imshow(image_array, cmap='Greys', interpolation='None')

    print("\n\nActual Value: ", y_test[5])
    
    return x_train, y_train, x_test, y_test
    #__________________________________________________________________________________________________________________________

    



```

```python

def AI_Accuracy(Y_pred, y_test):
    index = 0 
    scorecard = 0

    while index < 10:
        print("\n\nPrediction:", Y_pred[index])
        print("\tActual:", y_test[index])

        if Y_pred[index] == y_test[index]:
            scorecard += 1.00
            pass

        index += 1
        pass

    accuracy = scorecard / 10.00
    print("\n\n\n\nAccuracy: ", accuracy)

    return accuracy
```

```python
#2. Create Reservoir

x_train, y_train, x_test, y_test = MNIST_process_data()

from reservoirpy.nodes import Reservoir, Ridge, Input

source = Input()
reservoir = Reservoir(1400, sr=0.9, lr=0.5)
readout = Ridge(ridge=1e-6)

esn_model = [source >> reservoir, source] >> readout

esn_model = esn_model.fit(x_train[0:10000], y_train[0:10000]) #This trains the esn_model using the X_train as the input data and Y_train as the target data 
# Does the same thing as the previous program, just more condensed

print(reservoir.is_initialized, readout.is_initialized, readout.fitted)

#This runs the X data through the reservoir and produces output that should match the predicted future curve
Y_pred = esn_model.run(x_test[0:10])




```

```python

AI_Accuracy(Y_pred[0:10], y_test[0:10])
```
